```python
# datapoint
```

```python
data['Obs_Value'].hasnans
```

```python
dps = data[['Country_Code', 'Indicator_Code', 'Sex_Item_Code', 'Time', 'Obs_Value']].copy()
```

```python
dps.columns = ['country', 'concept', 'sex', 'year', 'value']
```

```python
dps['country'] = dps['country'].map(to_concept_id)
dps['sex'] = dps['sex'].map(to_concept_id)
dps['concept'] = dps['concept'].map(to_concept_id)
```

```python
gps = dps.groupby('concept').groups
```

```python
for k, idxs in gps.items():
    df = dps.ix[idxs]
    if np.all(df['sex'].isnull()):
        df_ = df[['country', 'year', 'value']].copy()
        path = '../ddf--datapoints--{}--by--country--year.csv'.format(k)
    else:
        assert not df['sex'].hasnans
        df_ = df[['country', 'sex', 'year', 'value']].copy()
        path = '../ddf--datapoints--{}--by--country--sex--year.csv'.format(k)
    
    df_ = df_.rename(columns={'value': k})
    
#     print(np.all(df_[['country', 'year']].duplicated()))
    
    df_.sort_values(by=['country', 'year']).to_csv(path, index=False, float_format='%g')
```
